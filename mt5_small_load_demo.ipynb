# mT5-small 加载示例

这个 notebook 只演示如何使用 `transformers` 加载 `mt5-small` 模型和分词器。

```python
# 如果本地还没有安装依赖，可以先执行这一格
# 正常已经安装过的话，这一格可以跳过

# !pip install transformers sentencepiece torch
```

```python
# 导入需要的库
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 这里使用 Google 的 mT5-small
model_name = "google/mt5-small"

print("准备加载模型:", model_name)
```

```python
# 加载分词器
# 第一次运行时会自动从 Hugging Face 下载到本地缓存
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)

# 加载模型
# 这是一个 encoder-decoder 架构，适合做文本到结构化输出的任务
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

print("模型和分词器加载完成")
```

```python
# 简单看一下 tokenizer 和模型的一些基本信息
print("tokenizer 类型:", type(tokenizer).__name__)
print("模型类型:", type(model).__name__)
print("词表大小:", tokenizer.vocab_size)
print("模型参数量(近似):", sum(p.numel() for p in model.parameters()))
```

```python
# 用一条中文输入做分词测试
text = "设计一个完成RC低通滤波功能的EDA原理图，包含输入VIN、电阻R1、输出OUT、电容C1和GND。"

encoded = tokenizer(text, return_tensors="pt")

print("输入文本:")
print(text)
print()
print("input_ids shape:", encoded["input_ids"].shape)
print("attention_mask shape:", encoded["attention_mask"].shape)
print("前20个 token id:", encoded["input_ids"][0][:20].tolist())
```

```python
# 再做一次最简单的生成测试
# 这里只是验证调用方式是否正常，不代表模型已经学会你的任务
generated = model.generate(
    **encoded,
    max_length=64
)

decoded = tokenizer.batch_decode(generated, skip_special_tokens=True)
print("模型原始生成结果:")
print(decoded[0])
```
